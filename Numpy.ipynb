Massiv elementlariga murojaat qilish

```python
arr = np.array([
    [10, 20, 30],
    [40, 50, 60],
    [70, 80, 90]
])
print(arr)
```

```python
print(arr[1,1])
```

```python
print(arr[1])
```

uchunchi ustunni olish

```python
print(arr[:, 2])
```

Boolen indexlash

```python
arr1 = np.arange(0, 56, 5)
print(arr1)
```

```python
print(arr1[arr1>10])
```

Boolen indexlashda AND operatori

```python
arr2 = np.array([8, 15, 3, 22, 10, 18, 5])
print(arr2)
```

```python
print(arr2[(arr2 > 10) & (arr2 < 20)])
```
